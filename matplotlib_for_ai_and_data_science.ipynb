# 🎨 1. Introduction to Matplotlib

***Matplotlib*** is the most popular data visualization library in Python. It enables you to transform raw data into clear, informative, and visually appealing charts. Whether you're exploring datasets, analyzing trends, or presenting insights, Matplotlib provides a wide range of plotting tools to make your data easier to understand.

It serves as the foundation for many other visualization libraries, such as ***Seaborn***, and is widely used in ***data science, machine learning, AI, research, and business analytics***.

In this notebook, you'll learn how to create, customize, and style a variety of plots—from simple line charts to advanced visualizations—using Matplotlib.

# 3. LINE CHART
A line chart is used to display data in form of 2D


## 3.1 Plotting a basic line chart

```python
ages = [18, 20, 22, 25, 28, 30, 35, 40, 45, 50, 55, 60]
salaries = [25000, 30000, 40000, 55000, 65000, 70000, 80000, 90000, 95000, 100000, 98000, 90000]
 
plt.plot(ages, salaries)
plt.show()
```

## 3.2 Adding a title

```python
plt.plot(ages, salaries)
plt.title("Age vs Salary")
plt.show()
```

## 3.3 Adding labels to x and y axes

```python
plt.plot(ages, salaries)

plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary (USD)")

plt.show()
```

## 3.4 Changing the font and resizing title/labels

```python
title_font   = {"fontname": "DejaVu Sans", "fontsize": 18, "fontweight": "bold", "color": "darkblue"}
label_font   = {"fontname": "DejaVu Sans", "fontsize": 13, "color": "darkgreen"}
 
plt.plot(ages, salaries)

plt.title("Age vs Salary", fontdict=title_font)
plt.xlabel("Age",    fontdict=label_font)
plt.ylabel("Salary (USD)", fontdict=label_font)

plt.show()
```

## 3.4 (continued) Editing tick marks

```python
x_ticks = [18, 25, 30, 35, 40, 45, 50, 55, 60]
y_ticks = [20000, 40000, 60000, 80000, 100000]
 
plt.plot(ages, salaries)

plt.title("Age vs Salary", fontdict=title_font)
plt.xlabel("Age",    fontdict=label_font)
plt.ylabel("Salary (USD)", fontdict=label_font)

plt.xticks(x_ticks, rotation=45, fontsize=10, color="grey")
plt.yticks(y_ticks, ["$20K", "$40K", "$60K", "$80K", "$100K"], fontsize=10, color="grey")

plt.show()
```

## 3.5 Adding a legend

```python
plt.plot(ages, salaries, label="Developer Salary")

plt.title("Age vs Salary", fontdict=title_font)
plt.xlabel("Age",    fontdict=label_font)
plt.ylabel("Salary (USD)", fontdict=label_font)

plt.xticks(x_ticks, rotation=45, fontsize=10, color="grey")
plt.yticks(y_ticks, ["$20K", "$40K", "$60K", "$80K", "$100K"], fontsize=10, color="grey")

plt.legend()

plt.show()
```

## 3.6 Changing line style

```python
#   color, linewidth, linestyle, marker, markersize, markeredgecolor
plt.plot(
    ages, salaries,
    label="Developer Salary",
    color="blue",
    linewidth=2,
    linestyle="--",       # options: '-', '--', '-.', ':'
    marker="o",           # dot at each data point
    markersize=7,
    markeredgecolor="red"
)

plt.title("Age vs Salary", fontdict=title_font)
plt.xlabel("Age",    fontdict=label_font)
plt.ylabel("Salary (USD)", fontdict=label_font)

plt.xticks(x_ticks, rotation=45, fontsize=10, color="grey")
plt.yticks(y_ticks, ["$20K", "$40K", "$60K", "$80K", "$100K"], fontsize=10, color="grey")

plt.legend()

plt.show()
```

## 3.6 (Continued) Changing line style using Shorthand

```python
plt.plot(ages, salaries, "o--b", lw=2, ms=7, mec="red", label="Developer Salary",)

plt.title("Age vs Salary", fontdict=title_font)
plt.xlabel("Age",    fontdict=label_font)
plt.ylabel("Salary (USD)", fontdict=label_font)

plt.xticks(x_ticks, rotation=45, fontsize=10, color="grey")
plt.yticks(y_ticks, ["$20K", "$40K", "$60K", "$80K", "$100K"], fontsize=10, color="grey")

plt.legend()

plt.show()
```

## 3.7 Adding a 2nd line

```python
ages2     = [18, 20, 22, 25, 28, 30, 35, 40, 45, 50, 55, 60]
salaries2 = [20000, 25000, 35000, 45000, 52000, 58000, 65000, 72000, 78000, 82000, 80000, 75000]
 
plt.plot(
    ages, salaries,
    label="Developer",
    color="blue", linewidth=2, linestyle="--",
    marker="o", markersize=7, markeredgecolor="red"
)
plt.plot(
    ages2, salaries2,
    label="Designer",
    color="green", linewidth=2, linestyle="-.",
    marker="s", markersize=7, markeredgecolor="orange"
)
plt.title("Age vs Salary", fontdict=title_font)
plt.xlabel("Age",    fontdict=label_font)
plt.ylabel("Salary (USD)", fontdict=label_font)

plt.xticks(x_ticks, rotation=45, fontsize=10, color="grey")
plt.yticks(y_ticks, ["$20K", "$40K", "$60K", "$80K", "$100K"], fontsize=10, color="grey")

plt.legend()
plt.show()
```

## 3.8 Resizing a chart

```python
plt.figure(figsize=(12, 6), dpi=100)   # width=12in, height=6in, 100 dots-per-inch
 
plt.plot(ages, salaries, label="Developer", color="blue",  linewidth=2,
         linestyle="--", marker="o", markersize=7, markeredgecolor="red")
plt.plot(ages2, salaries2, label="Designer", color="green", linewidth=2,
         linestyle="-.", marker="s", markersize=7, markeredgecolor="orange")

plt.title("Age vs Salary", fontdict=title_font)
plt.xlabel("Age",    fontdict=label_font)
plt.ylabel("Salary (USD)", fontdict=label_font)

plt.xticks(x_ticks, rotation=45, fontsize=10, color="grey")
plt.yticks(y_ticks, ["$20K", "$40K", "$60K", "$80K", "$100K"], fontsize=10, color="grey")

plt.legend()
plt.show()
```

## 3.9 Saving a chart

```python
plt.figure(figsize=(12, 6), dpi=100)
 
plt.plot(ages, salaries, label="Developer", color="blue",  linewidth=2,
         linestyle="--", marker="o", markersize=7, markeredgecolor="red")
plt.plot(ages2, salaries2, label="Designer", color="green", linewidth=2,
         linestyle="-.", marker="s", markersize=7, markeredgecolor="orange")

plt.title("Age vs Salary", fontdict=title_font)
plt.xlabel("Age",    fontdict=label_font)
plt.ylabel("Salary (USD)", fontdict=label_font)

plt.xticks(x_ticks, rotation=45, fontsize=10, color="grey")
plt.yticks(y_ticks, ["$20K", "$40K", "$60K", "$80K", "$100K"], fontsize=10, color="grey")
plt.legend()

plt.tight_layout()                       # prevent labels being clipped
plt.savefig("age_vs_salary.png")         # saves to current directory

plt.show()
```

# 4. BAR CHART

```python
languages = ["Python", "JavaScript", "Java", "C++", "Go", "Rust"]
popularity = [30, 25, 20, 12, 8, 5]
```

## 4.1 Plotting a basic bar chart

```python
#         (size changed the same way as line chart: figure(figsize=(), dpi=))
plt.figure(figsize=(10, 6), dpi=100)

bars = plt.bar(languages, popularity)

plt.title("Programming Language Popularity",
          fontdict={"fontsize": 16, "fontweight": "bold"})

plt.xlabel("Language", fontdict={"fontsize": 13})
plt.ylabel("Popularity (%)", fontdict={"fontsize": 13})

plt.show()
```

## 4.2 Setting bar patterns

```python
#   hatch patterns: '/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*'
hatch_patterns = ['/', '\\', '|', '-', '+', 'x']
colors         = ['steelblue', 'tomato', 'gold', 'mediumseagreen', 'orchid', 'coral']
 
plt.figure(figsize=(10, 6), dpi=100)
bars = plt.bar(languages, popularity, color=colors, edgecolor="black", linewidth=1.2)
 
# Apply a different hatch pattern to each bar
for bar, pattern in zip(bars, hatch_patterns):
    bar.set_hatch(pattern)
 
plt.title("Programming Language Popularity (with Hatch Patterns)",
          fontdict={"fontsize": 15, "fontweight": "bold"})
plt.xlabel("Language", fontdict={"fontsize": 13})
plt.ylabel("Popularity (%)", fontdict={"fontsize": 13})

plt.tight_layout()
plt.savefig("language_popularity_bar.png")
plt.show()
```

## 5.2 Example 2 - Histogram
Explore the distribution and frequency of numerical data with a histogram.

### 5.2.1 Initializing Data
Generate sample data to analyze its distribution.

```python
fifa = pd.read_csv("/kaggle/input/datasets/joebeachcapital/fifa-players/fifa_2022_datasets/players_22.csv", low_memory=False)
fifa.head()
```

### 5.2.2 Making Histogram
Plot a histogram to visualize how values are distributed across intervals.

```python
fig, ax = plt.subplots(figsize=(8, 5), dpi=100)

counts, bins, patches = ax.hist(
    fifa.overall, 
    bins=20, 
    color="#4A90E2",      # Modern soft blue
    edgecolor="white",    # Clean separation between bars
    rwidth=0.9,           # 10% gap between bars
    alpha=0.85            # Subtle transparency
)

ax.set_title("Distribution of Player Skills in FIFA 2022", fontsize=14, pad=15, weight="bold", color="#333333")
ax.set_xlabel("Skill Level", fontsize=11, labelpad=10, color="#555555")
ax.set_ylabel("Number of Players", fontsize=11, labelpad=10, color="#555555")

# 5. Clean up borders (Spines)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_color("#cccccc")
ax.spines["bottom"].set_color("#cccccc")

# 6. Add a subtle background grid
ax.grid(axis="y", linestyle="--", alpha=0.5, zorder=0)
ax.set_axisbelow(True) # Ensures grid lines sit behind the bars

# 7. Adjust tick parameters
ax.tick_params(axis="both", colors="#555555", labelsize=10)

plt.tight_layout()
plt.savefig("distribution_of_player_skills.png", dpi=300)

plt.show()
```

## 5.3 Example 3 - Pie Chart
Illustrate the proportion of each category as parts of a whole using a pie chart.

```python
left = fifa.loc[fifa.preferred_foot == "Left"].count().iloc[0]
right = fifa.loc[fifa.preferred_foot == "Right"].count().iloc[0]

# 1. Data definition
labels = ['Left', 'Right']
sizes = [left, right]

# 2. Modern, cohesive color palette (Muted/Pastel)
colors = ['#8db580', '#7ca1cc']

# 3. Highlight a specific slice smoothly (optional)
explode = (0, 0.05)  # slightly slide out the 2nd slice

# 4. Create the plot
fig, ax = plt.subplots(figsize=(6, 6), facecolor='#fafafa')

wedges, texts, autotexts = ax.pie(
    sizes, 
    explode=explode, 
    labels=labels, 
    autopct='%1.1f%%',
    startangle=90,             # Starts top-center
    colors=colors, 
    pctdistance=0.75,          # Moves percentages closer to center
    wedgeprops={
        'linewidth': 1,        # White border thickness
        'edgecolor': 'white'   # Creates clean division
    },
    textprops={'fontsize': 11, 'color': '#333333'} # Dark gray text
)

# 5. Styling text inside the slices separately
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_weight('bold')
    autotext.set_fontsize(10)

# 6. Final aesthetics
ax.set_title('Foot Prefrence of FIFA Players', fontsize=14, pad=20, weight='bold', color='#222222')
ax.axis('equal')  # Ensures the pie is a perfect circle

plt.tight_layout()
plt.savefig("foot_prefrence_of_players.png", dpi=300)

plt.show()
```

## 5.4 Example 4 - Box Plot
Summarize data distribution and detect outliers using a box plot.

```python
labels = ['Overall Rating', 'Potential Rating']

fig, ax = plt.subplots(figsize=(8, 6), dpi=100)
ax.set_facecolor('#F9FBFD')  # Soft background color
ax.grid(True, linestyle='--', alpha=0.5, color='#CCCCCC', zorder=0)

box_props = dict(linewidth=2.0, edgecolor='#2C3E50')
whisker_props = dict(linewidth=1.5, linestyle='-', color='#7F8C8D')
cap_props = dict(linewidth=1.5, color='#7F8C8D')
median_props = dict(linewidth=2.5, color='#E74C3C')
flier_props = dict(marker='o', markerfacecolor='#E74C3C', markersize=6,
                   markeredgecolor='none', alpha=0.6)

bplot = ax.boxplot([fifa.overall, fifa.potential], 
                   tick_labels=labels, 
                   patch_artist=True, 
                   widths=0.55,
                   zorder=3,
                   boxprops=box_props,
                   whiskerprops=whisker_props,
                   capprops=cap_props,
                   medianprops=median_props,
                   flierprops=flier_props)

colors = ['#5DADE2', '#AF7AC5']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.85)

ax.set_title('Distribution Analysis across Rates', fontsize=14, pad=15, 
             fontweight='bold', color='#2C3E50')
ax.set_ylabel('FIFA Overall Rating', fontsize=11, labelpad=10, color='#34495E')

for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)
ax.spines['left'].set_color('#BDC3C7')
ax.spines['bottom'].set_color('#BDC3C7')

ax.tick_params(axis='both', colors='#34495E', labelsize=10)

plt.tight_layout()
plt.savefig("distribution_analysis_across_rates.png", dpi=300)

plt.show()
```
