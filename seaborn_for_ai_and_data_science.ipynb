# 1. Why Seaborn?

Working with Matplotlib directly can be verbose. Seaborn wraps common statistical plotting patterns into clean, one-line calls while producing publication-quality figures out of the box.

| Feature | Matplotlib | Seaborn |
|---|---|---|
| Lines of code for a styled plot | Many | Few |
| Built-in themes | None | 5 built-in |
| Pandas integration | Partial | Native |
| Statistical plots | Manual | Built-in |
| Color palettes | Manual | Rich & automatic |

# 4. Scatter Plot
To make a [**scatter plot**](https://seaborn.pydata.org/generated/seaborn.scatterplot.html) in Python using Seaborn, use the `sns.scatterplot()` function.

Seaborn allows you to display up to three additional variables by changing point attributes:
- `hue`: Groups data by color using a third variable.
- `style`: Changes the marker shapes based on a variable.
- `size`: Alters the point radius based on a variable.

```python
sns.scatterplot(penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", style="sex")
```

# 5. Styling in Seaborn

## 5.1 The `set_style()`

Seaborn features five background styles to suit different presentation needs:
- `darkgrid`: The default choice. It uses a grey background with white grid lines.
- `whitegrid`: A clean, white background with subtle grey grid lines.
- `dark`: A solid grey background without any grid lines.
- `white`: A completely crisp, plain white background without grid lines.
- `ticks`: Similar to the white style, but adds small tick marks directly onto the axis lines

## 5.2 The `set_context()`
In Seaborn, `sns.set_context()` is a function that quickly scales the visual elements of your plot (like label sizes, line widths, and font sizes) to fit different presentation mediums.

### 4 Pre-set Contexts (from smallest to largest scale):
- `paper`: Default size; optimized for academic documents and reports.
- `notebook`: Slightly larger; optimized for Jupyter notebooks.
- `talk`: Bold and large; optimized for presentation slides.
- `poster`: Extremely large; optimized for large-scale exhibition posters.

```python
sns.scatterplot(penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
sns.set_context("notebook")
```

# 6. Palettes in Seaborn

A [palette](https://seaborn.pydata.org/tutorial/color_palettes.html) in Seaborn is a set of colors used to make data graphs easy to read. It controls the colors of bars, lines, and dots. Seaborn offers many built-in [Color Palette](https://seaborn.pydata.org/tutorial/color_palettes.html) options for different types of data.

## 6.1 Palette Types

* **Qualitative (Categorical):** For distinct groups with no order (e.g., countries, categories).
> *Popular:* `deep` (default), `pastel`, `colorblind` (accessible).


* **Sequential (Numeric/Grayscale):** For values moving from low to high (e.g., age, count).
> *Popular:* `Blues`, `rocket`, `mako` (append `_r` to reverse, like `blues_r`).


* **Diverging (Two Extremes + Midpoint):** For data with a critical middle point (e.g., temperature changes, correlations from -1 to 1).
> *Popular:* `coolwarm`, `vlag`, `RdBu`.

```python
sns.scatterplot(penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", palette="rocket")
```

<font face="Consolas"><font>
## 6.2 Palette Styles

* **`"deep"`**: Default palette, rich and balanced
* **`"muted"`**: Softer colors, good for many categories
* **`"bright"`**: High-contrast, bold colors
* **`"pastel"`**: Soft, pale tones
* **`"dark"`**: Darker tones, easy to distinguish
* **`"colorblind"`**: Optimized for colorblind accessibility

```python
sns.scatterplot(penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", palette="dark")
```

<font face="Consolas"><font>
## 6.3 ColorBrewer Qualitative Sets

* **`"Set1"`**: Bold, high-contrast, heavily saturated colors (up to 9 categories).
* **`"Set2"`**: Muted, softer, mid-tone pastel colors (up to 8 categories).
* **`"Set3"`**: Light, bright pastel colors, ideal for larger groups (up to 12 categories).

```python
sns.scatterplot(penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", palette="Set2")
```

# 7. Strip Plots in Seaborn

A [**strip plot**](https://seaborn.pydata.org/generated/seaborn.stripplot.html) is basically a scatter plot where one of the variables is categorical. It draws a single point for each observation along a categorical axis, showing the underlying distribution of your data.

## 7.1 Quick Syntax


```python
# The standard format
sns.stripplot(data=penguins, x="species", y="body_mass_g")
```

<font face="Consolas"><font>
## 7.2 Key Adjustments

Because points can easily stack on top of each other and hide the data density, you will almost always want to use these parameters:

* **`jitter=True`**: Adds a tiny bit of random noise/space horizontally so overlapping points spread out and become visible.
* **`hue`**: Splits your categories further by coloring points based on a second categorical variable.
* **`dodge=True`**: When using `hue`, this separates the colored strips into their own distinct columns instead of mixing them together.

```python
# Fully optimized example
sns.stripplot(data=penguins, x="species", y="body_mass_g", jitter=True, hue="island", dodge=True)
```

This makes it easy to see the distribution of data across different categories.

```python
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue="island")
```

# 9. Histogram in Seaborn

# 10. Regression Plot in Seaborn

# 11. Line Plot in Seaborn

A [**Line Plot**](https://seaborn.pydata.org/generated/seaborn.lineplot.html) is used to visualize **trends or changes over an ordered variable**, such as time, sequence, or continuous data. It connects data points with lines, making it easy to identify patterns and trends.

```python
sns.lineplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
```

# 12. Joint Plot in Seaborn

A [**Joint Plot**](https://seaborn.pydata.org/generated/seaborn.jointplot.html) combines **two visualizations in one figure**:

* 📊 A central plot showing the relationship between two variables.
* 📈 Marginal plots (histograms, KDEs, etc.) showing the distribution of each variable individually.

It is useful for exploring both correlation and distributions at the same time.

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
```

# 13. Bar Plot in Seaborn

A [Bar Plot](https://seaborn.pydata.org/generated/seaborn.barplot.html) is used to compare the average (or another aggregated statistic) of a numerical variable across different categories. By default, Seaborn displays the mean and a confidence interval (error bar) for each category.

```python
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette="rocket")
```

# 14. Count Plot in Seaborn

A [Count Plot](https://seaborn.pydata.org/generated/seaborn.countplot.html) is used to display the number of observations (frequency) in each category of a categorical variable. Unlike a bar plot, it counts the occurrences rather than calculating an aggregate such as the mean

```python
sns.countplot(data=penguins, x="species")
```

# 15. Box Plot in Seaborn

A [**Box Plot**](https://seaborn.pydata.org/generated/seaborn.boxplot.html) visualizes the distribution of numerical data using the five-number summary:

* Minimum
* First Quartile (Q1)
* Median (Q2)
* Third Quartile (Q3)
* Maximum

It is especially useful for detecting **outliers**, comparing distributions, and understanding the spread of data.

```python
sns.boxplot(data=penguins, x="species", y="body_mass_g", hue="sex")
```

---
<font face="Consolas"><font>
# 16. Violin Plot in Seaborn

A [**Violin Plot**](https://seaborn.pydata.org/generated/seaborn.violinplot.html) combines a **Box Plot** and a **Kernel Density Estimate (KDE)** to show both the distribution and the probability density of numerical data across categories. It provides more information about the shape of the distribution than a standard box plot.

```python
sns.violinplot(data=penguins, x="species", y="body_mass_g", hue="sex", split=True)
```

<font face="Consolas"><font>
# 17. KDE Plot in Seaborn

A [**KDE**](https://seaborn.pydata.org/generated/seaborn.kdeplot.html) (**Kernel Density Estimation**) Plot visualizes the probability density of a numerical variable using a smooth curve. It provides a continuous estimate of the data distribution, making it a smoother alternative to a histogram.

```python
sns.kdeplot(data=penguins, x="body_mass_g", hue="species", fill=True)
```

# 18. Heatmap in Seaborn

A [**Heatmap**](https://seaborn.pydata.org/generated/seaborn.heatmap.html) visualizes data in a **matrix format**, where values are represented by colors. It is commonly used to display **correlation matrices**, missing values, and other tabular data patterns.

```python
sns.heatmap(data=penguins.corr(numeric_only=True), annot=True, cmap="coolwarm")
```

# 19. Rug Plot in Seaborn

A [**Rug Plot**](https://seaborn.pydata.org/generated/seaborn.rugplot.html) displays **individual data points as small ticks** along an axis. It provides a simple view of where observations are located and is often combined with **histograms** or **KDE plots** to better understand the distribution of data.

# 20. Pair Plot in Seaborn

A [**Pair Plot**](https://seaborn.pydata.org/generated/seaborn.pairplot.html) creates a grid of plots showing **pairwise relationships** between multiple numerical variables in a dataset. It displays scatter plots for variable pairs and **histograms** (or KDE plots) on the diagonal, making it an excellent tool for exploratory data analysis (EDA).

<font face="Consolas"><font>
## 20.1 Basic Pair Plot
To build a default pair plot, simply pass your tidy, long-form DataFrame directly into the function

```python
sns.pairplot(data=penguins)
```

<font face="Consolas"><font>
## 20.2 Segmenting Data by Categories (`hue`)
You can use the `hue` argument to map a categorical variable to distinct colors. This helps you spot distinct clusters or patterns across classifications.
* When `hue` is provided, the diagonal subplots automatically change from histograms to **layered Kernel Density Estimate (KDE)** plots

```python
sns.pairplot(data=penguins, hue="species", palette="rocket")
```

<font face="Consolas"><font>
## 20.3 Customizing Plot Types (`kind` & `diag_kind`)
You can modify the display structure of both off-diagonal and diagonal subplots.
* `kind` (Off-diagonal): Choose between `'scatter'`, `'kde'`, `'hist'`, or `'reg'` (adds a linear regression line).
* `diag_kind` (Diagonal): Choose between `'hist'`, `'kde'`, or `None`.

```python
sns.pairplot(data=penguins, kind="reg", diag_kind="kde")
```

<font face="Consolas"><font>
## 20.4 Creating a Clean "Corner" Layout
Because pair plot grids are symmetrical across the diagonal, the top-right triangle mirrors the bottom-left triangle. You can pass `corner=True` to hide the redundant upper triangle, making the layout cleaner and easier to read

```python
sns.pairplot(data=penguins, hue="species", palette="rocket", corner=True)
```

# 21. PairGrid in Seaborn

A [PairGrid](https://seaborn.pydata.org/generated/seaborn.PairGrid.html) is a customizable version of a Pair Plot. It creates a grid of pairwise relationships between variables, allowing you to specify different plot types for the upper triangle, lower triangle, and diagonal.

It is useful when you need more flexibility than `pairplot()` provides.

<font face="Consolas"><font>
## 21.1 Basic Setup
Using `PairGrid` is a two-step process: initialize the empty grid infrastructure with your data, then apply plotting functions to it.

```python
# 1. Initialize the grid
grid = sns.PairGrid(data=penguins)

# 2. Map a plot type to all axes
grid.map(sns.scatterplot)
```

<font face="Consolas"><font>
## 21.2 Advanced Customization
You can control what displays in different quadrants using specific map methods.

For example, you can plot histograms on the diagonal, regression lines on the lower triangle, and kernel density estimates (KDE) on the upper triangle.

```python
g = sns.PairGrid(penguins, hue="species", palette="rocket")

# Set univariate plots on the diagonal
g.map_diag(sns.histplot, element="step")

# Set bivariate plots on the lower triangle
g.map_lower(sns.scatterplot)

# Set bivariate plots on the upper triangle
g.map_upper(sns.kdeplot, alpha=0.5)

# Add a legend for the hue variable
g.add_legend()
```
